```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(100) + (np.random.random(100) * 5)
y = np.arange(100) + (np.random.random(100) * 5)
z = np.arange(100)
```

```python
fig, ax = plt.subplots()
s =ax.scatter(x, y, c=z, s=z, cmap=plt.get_cmap('gist_earth_r'),
             alpha=0.5, edgecolor='k')
plt.colorbar(s)
```
